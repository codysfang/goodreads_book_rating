```python
import numpy as np
import pandas as pd

np.random.seed(409782)
```

```python
train = pd.read_csv(f"train_final_unscaled.csv")
test = pd.read_csv(f"test_final_unscaled.csv")

train_y = train.pop("rating_label")
train_X = train
test_X = test

train_X
```

```python
from sklearn.model_selection import RepeatedKFold, KFold
from sklearn import metrics

def report(a, b):
    print("New Split reports")
    reports = [
                metrics.accuracy_score(a, b), 
                metrics.precision_score(a, b, average="macro"),
                metrics.recall_score(a, b, average="macro"),
                metrics.f1_score(a, b, average="macro")
               ]
    return reports
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

train_split_X = train_X.iloc[:18350, :]
train_split_y = train_y[:18350]
test_split_X = train_X.iloc[18351:, :]
test_split_y = train_y[18351:]


logis = LogisticRegression(random_state=10987, max_iter=10000)
logis.fit(train_split_X, train_split_y)
pred_y = logis.predict(test_split_X)
```

```python
report(pred_y, test_split_y)
```
